```python
from sklearn.preprocessing import normalize

# load csv file 
def loadData(path):   
    print("Data loading...")
    
    data = []
    
    # Read csv file from directory
    data = genfromtxt(path, delimiter=',' ,encoding='UTF8')

                
    # Remove colums and row
    Data = data[1:,3:7]
   
        
    # Convert string to float64
    Data.astype('float64')
 
   
    # Fill NaN data to 0   
    Data[np.isnan(Data)] = 0
    
    for i in range(Data.shape[0]):
        for j in range(Data.shape[1]):
            if Data[i][j] == 0:
                Data[i][j] = (Data[i-1][j] + Data[i+1][j])/2
      

    # Dataframe
    DF = pd.DataFrame(Data, columns = ['기온', '이슬점온도', '풍속', '해면기압'])
    
    Data = normalize(Data, axis=0, norm='max')
    
    return Data, DF
```

```python
Jeju = './Weather/제주(140801~140831).csv'
Jeju_array, Jeju_dataFrame = loadData(Jeju)
```
